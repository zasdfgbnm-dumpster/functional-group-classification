```python
import caffe
import h5py

modelfn = 'lenet_train_test.prototxt'
modelwfn = 'irnet-result_iter_35000.caffemodel'
net = caffe.Net(modelfn,modelwfn,caffe.TEST)

dbsize = 3000
batchsize = 60
testdb = h5py.File('data-3.h5')
countpp = 0
countpn = 0
countnp = 0
countnn = 0
count = 0
for i in range(dbsize/batchsize):
    batch_data = testdb['data'][i*batchsize:(i+1)*batchsize]
    batch_label_raw = testdb['label'][i*batchsize:(i+1)*batchsize]
    batch_label = [int(e[0][0][0]) for e in batch_label_raw]
    net.blobs['data'].data[...] = batch_data
    net.forward()
    predicted_label_raw = net.blobs['ip2'].data
    predicted_label = [ 0 if x[0]>x[1] else 1 for x in predicted_label_raw]
    for e,p in zip(batch_label,predicted_label):
        # print 'batch:',i,'\texpected:',e,'\tgot:',p
        count += 1
        if e and p:
            countpp += 1
            continue
        if e and not p:
            countpn += 1
            continue
        if not e and p:
            countnp += 1
            continue
        if not e and not p:
            countnn += 1
            continue

print 'overall accuracy:',100.0*(countpp+countnn)/count,'%'
print '+ accuracy:',100.0*countpp/(countpp+countpn),'%'
print '- accuracy:',100.0*countnn/(countnp+countnn),'%'
```
